```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"

import pandas as pd 
import math
from sklearn.linear_model import LinearRegression

def gradient_descent(x, y):
    n = x.size 
    m = b = error = 0
    learning_rate = 0.0002
    
    iterations = 100000
    for i in range(iterations):
        yhat = m * x + b
        error_previous = error 
        error = 1/n * sum((y - yhat) ** 2)
        # print("iteration = {}, m = {}, b = {}, error = {}".format(i, m, b, error))
        if math.isclose(error, error_previous, rel_tol = 1e-20): 
            break
        
        bd = -2/n * sum(y - yhat) 
        md = -2/n * sum(x * (y - yhat))
        m -= learning_rate * md
        b -= learning_rate * bd 
    return m, b
    

if __name__ == "__main__":
    df = pd.read_csv("~/code/py/ML/3_gradient_descent/Exercise/test_scores.csv")
    x = df['math']
    y = df['cs']
    m, b = gradient_descent(x, y)
    m, b 

    reg = LinearRegression()
    reg.fit(df[['math']], y) 
    reg.coef_, reg.intercept_
```

```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = 'all' 

import kagglehub
path = kagglehub.dataset_download("fedesoriano/heart-failure-prediction")

import pandas as pd
df = pd.read_csv(path + '/heart.csv') 
df.head()

y = df['HeartDisease']
df.drop('HeartDisease', axis = 'columns', inplace=True) 
df.head()

import numpy as np
cols = df.select_dtypes(include = [np.number]).columns
cols
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler().set_output(transform="pandas")
df[cols] = scaler.fit_transform(df[cols])
df.head()

df['Sex'].unique()
df['Sex'] = df['Sex'].map({'M': 0, 'F': 1}) 
df['ExerciseAngina'].unique()
df['ExerciseAngina'] = df['ExerciseAngina'].map({'N': 0, 'Y': 1})
df['RestingECG'].unique()
df['ST_Slope'].unique()
df['ChestPainType'].unique()
df = pd.get_dummies(df, drop_first=True)
df.head()
X = df

from sklearn.model_selection import cross_val_score
from sklearn.svm import SVC
scores = cross_val_score(SVC(), X, y, cv = 5) 
scores.mean()
from sklearn.tree import DecisionTreeClassifier
scores = cross_val_score(DecisionTreeClassifier(), X, y, cv = 5) 
scores.mean()

from sklearn.ensemble import BaggingClassifier
# model = BaggingClassifier(estimator=SVC(), n_estimators=50, max_samples=0.8, oob_score=True)
model = BaggingClassifier(estimator=SVC(), n_estimators=50, oob_score=True)
scores = cross_val_score(model, X, y, cv = 5) 
scores.mean()
# model = BaggingClassifier(estimator=DecisionTreeClassifier(), n_estimators=50, max_samples=0.8, oob_score=True)
model = BaggingClassifier(estimator=DecisionTreeClassifier(), n_estimators=50, oob_score=True)
scores = cross_val_score(model, X, y, cv = 5) 
scores.mean()
```
